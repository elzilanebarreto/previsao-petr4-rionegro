```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
```

```python
# Carregando o Dataset
dados = pd.read_csv('petr4.csv')
```

```python
dados.head()
```

```python
# Exemplo de como garantir que o índice seja datetime
dados['data_pregao'] = pd.to_datetime(dados['data_pregao'], format='%Y-%m-%d', errors='coerce')

# Definir 'data_pregao' como índice
dados.set_index('data_pregao', inplace=True)
```

```python
dados.head()
```

```python
# Selecionar apenas a coluna 'preco_ultimo_negocio' 
dados = dados[['preco_ultimo_negocio']]

# Completar os dias faltantes 
dados = dados.resample('D').ffill()
```

```python
dados.head()
```

```python
# Visualizar a série temporal
plt.figure(figsize=(7,5))
plt.plot(dados.index, dados['preco_ultimo_negocio'])
plt.xlabel('Ano')
plt.ylabel('Preço do Último Negócio')
plt.title('Preço de Fechamento da PETR4 ao Longo do Tempo')
plt.show()
```

```python
# Teste de Augmented Dickey-Fuller (ADF)
resultado = adfuller(dados['preco_ultimo_negocio'])
print(f"Estatística ADF: {resultado[0]}")
print(f"p-Valor: {resultado[1]}")

print('Valores Criticos:')
for chave, valor in resultado[4].items():
    print(f"{chave}: {valor}")
```

# Hipótese

## Comparação da Estatística ADF com os Valores Críticos:
A estatística ADF (-2.17580) é maior (menos negativa) do que todos os valores críticos:

    Nível de 1%: -3.43617
    Nível de 5%: -2.86411
    Nível de 10%: -2.56814
    
Para rejeitar a hipótese nula (de que a série é não estacionária), a estatística ADF deve ser menor (mais negativa) do que o valor crítico no nível de significância escolhido. Como a estatística ADF não é menor que nenhum dos valores críticos (1%, 5% ou 10%), não podemos rejeitar a hipótese nula com base nesse critério.
### p-valor:
    O p-valor é 0.21514, que é maior do que os níveis de significância comuns (0.01, 0.05 ou 0.10).
    Para rejeitar a hipótese nula, o p-valor deve ser menor que o nível de significância escolhido (ex.: 0.05). Como o p-valor (0.21514) é maior que 0.05 (e até 0.10), não há evidência estatística suficiente para rejeitar a hipótese nula.
## Conclusão:
Com base no p-valor (0.21514 > 0.05) e na comparação da estatística ADF com os valores críticos, não rejeitamos a hipótese nula. Isso indica que a série temporal é não estacionária.
Isso sugere que a série pode apresentar tendências, sazonalidade ou outras variações sistemáticas ao longo do tempo, o que impede que ela tenha média e variância constantes.

```python
# Calcule a média móvel (média móvel simples) com uma janela de 12 meses
rolling_mean = dados['preco_ultimo_negocio'].rolling(window=12).mean()

# Visualize a série temporal original e o componente de tendência
plt.figure(figsize=(7, 5))
plt.plot(dados.index, dados['preco_ultimo_negocio'], label='Original')
plt.plot(dados.index, rolling_mean, color='red', label='Tendência (Média Móvel)')
plt.xlabel('Ano')
plt.ylabel('Último Preço')
plt.title('Análise de Tendência: Último Preço de Negócio')
plt.legend()
plt.show()
```

### Há uma tendência crescente, pois o dado do último preço de fechamento de negócio da PETR4 no início do ano de 2020 é menor do que o último dado disponível em 2024.

```python
# Calcule o desvio padrão móvel com uma janela de 12 períodos
rolstd = dados['preco_ultimo_negocio'].rolling(window=12).std()

# Visualize a série temporal original e o desvio padrão móvel
plt.figure(figsize=(7, 5))
plt.plot(dados.index, dados['preco_ultimo_negocio'], label='Preço Original')
plt.plot(dados.index, rolstd, color='red', label='Desvio Padrão Móvel')
plt.xlabel('Ano')  # Ajuste para 'Data' ou similar, dependendo do índice de petr4
plt.ylabel('Preço Último Negócio')
plt.title('Análise da Volatilidade: Preço PETR4')
plt.legend()
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
# Gerando o gráfico de ACF (Função de Autocorrelação:medir o quanto 
# os valores de uma série estão correlacionados com eles mesmos em lags)
plt.figure(figsize=(7,5))
plot_acf(dados['preco_ultimo_negocio'], lags=50)
plt.xlabel('Lag')
plt.ylabel('Autocorrelação')
plt.title('Gráfico da Função de Autocorrelação (ACF)')
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
# Calculando o PACF (Função de Autocorrelação Parcial: mede a correlação
# entre os valores de uma série temporal em dois pontos no tempo)
plt.figure(figsize=(7, 5))
plot_pacf(dados['preco_ultimo_negocio'], lags =50)
plt.xlabel('Lag')
plt.ylabel('PACF')
plt.title('Função de Autocorrelação Parcial (PACF)')
plt.show()
```

```python
dados['diff1'] = dados['preco_ultimo_negocio'].diff()
plt.figure(figsize=(7, 5))
plt.plot(dados.index, dados['diff1'], label='Série Diferenciada (1ª Diferenciação)', color='orange')
plt.xlabel('Ano')
plt.ylabel('Diferença de Preço (R$)')
plt.title('Série Temporal após 1ª Diferenciação')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
plot_acf(dados['diff1'].dropna(), lags=10)
plt.xlabel('Lag')
plt.ylabel('Autocorrelação')
plt.title('ACF - Série Diferenciada')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
plot_pacf(dados['diff1'].dropna(), lags=10)
plt.xlabel('Lag')
plt.ylabel('Autocorrelação Parcial')
plt.title('PACF - Série Diferenciada')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(dados['preco_ultimo_negocio'], model='multiplicative', period=12)
plt.figure(figsize=(7, 5))
plt.subplot(411)
plt.plot(dados['preco_ultimo_negocio'], label='Original')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(decomposition.trend, label='Tendência')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Sazonalidade')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(decomposition.resid, label='Resíduo')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Calcule a Média Móvel Simples (SMA) com uma janela de 12 meses
sma = dados['preco_ultimo_negocio'].rolling(window=12).mean()

# Visualize a série original e a SMA
plt.figure(figsize=(8, 5))
plt.plot(dados.index, dados['preco_ultimo_negocio'], label='Original')
plt.plot(dados.index, sma, color='red', label='SMA (12 meses)')
plt.xlabel('Ano')
plt.ylabel('Preço Último Negócio')
plt.title('Análise de Tendência: SMA de 12 Meses no Preço')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Ajustar o modelo ARIMA(1,1,1) na série original 
model = ARIMA(dados['preco_ultimo_negocio'],order=(1, 1, 1))
fitted_model = model.fit()

# Fazer previsões para os próximos 12 meses
predictions = fitted_model.predict(start=dados.index[-1],end=dados.index[-1]+pd.DateOffset(months=12))

# Plotar a série original e as previsões
plt.figure(figsize=(8, 5))
plt.plot(dados['preco_ultimo_negocio'], label='Original')
plt.plot(predictions, color='red', label='Previsões')
plt.xlabel('Ano')
plt.ylabel('Preço da PETR4')
plt.title('Previsão do Preço da PETR4 com ARIMA')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
model = SARIMAX(dados['preco_ultimo_negocio'], order=(1, 1, 1), seasonal_order=(0, 1, 0, 12))
fitted_model = model.fit(disp=True)

# Fazer previsões
predictions = fitted_model.predict(start=dados.index[-1], end=dados.index[-1] + pd.DateOffset(months=12))

# Visualizar a série original e as previsões
plt.figure(figsize=(10, 6))
plt.plot(dados['preco_ultimo_negocio'], label='Original')
plt.plot(predictions, color='red', label='Predictions')
plt.xlabel('Ano')
plt.ylabel('Preço Último Negócio')
plt.title('Previsão do Preço Último Negócio usando SARIMA')
plt.legend()
plt.show()
```
